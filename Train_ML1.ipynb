# House price Prediction procject ( dự án  : Dự đoán giá nhà )
* Goal : predict the sales price for each house 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data_df = pd.read_csv("train.csv")
data_df.head()
```

```python
# ta thấy Id và số thứ tự bị trùng ( cần 1 cái là được )
data_df = pd.read_csv("train.csv",index_col="Id")
data_df.head()
```

```python
# Có tổng cộng 80 cột thì ta cần xem gồm những cột nào : 
data_df.columns
```

```python
# Choose features to train ML model ( chọn ra các cột cần thiết để train mô hình ML)
```

```python
features = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
```

### 3 Spliting dataset into X and Y 

```python
x = data_df[features]
x
```

```python
y = data_df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split 
np.random.seed(0)
Xtrain,Xtest,Ytrain,Ytest =train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=0)
Xtrain 
```

```python
Xtest
```

```python
Ytrain
```

```python
Ytest
```

### 4 Train Model :

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=1)
# dt là 1 mô hình rỗng , giờ ta cần đưa dữ liệu : 
dt.fit(Xtrain,Ytrain)
```

```python
# ta đã có được 1 mô hình đơn giản , giờ ta tiến hành dự đoán : 
y_predict = dt.predict(Xtest.head())
y_predict
```

```python
pd.DataFrame({'ytest':Ytest.head() , 'ypredict':y_predict})
```

```python
from sklearn.ensemble import RandomForestRegressor , GradientBoostingRegressor
```

```python
rf_model=RandomForestRegressor(random_state=1) # Mô hình này dùng để test hết dữ liệu phần Xtest để dự đoán 
```

```python
rf_model.fit(Xtrain,Ytrain)
```

```python
rf_predict=rf_model.predict(Xtest)
rf_predict[:5]
```

```python
# Bây giờ chỉ lấy dữ liệu của 1 ngôi nhà duy nhất để đưa ra dự đoán : 
Xtest
```
